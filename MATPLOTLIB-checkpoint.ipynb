```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"r")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.title(" Graph of X axis vs Y axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(x,y,"g*-")
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.9])
axes.plot(x,y,"g") 
axes.set_xlabel("X LABEL")
axes.set_ylabel("Y LABEL")
axes.set_title("title for graph between x and y")
```

```python
fig
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.9])
axes2=fig.add_axes([0.2,0.6,0.4,0.3])

axes1.plot(x,y,"g") 
axes1.set_xlabel("X1 LABEL")
axes1.set_ylabel("Y1 LABEL")
axes1.set_title("title for graph1 between x and y")

axes2.plot(x,y,"r") 
axes2.set_xlabel("X2 LABEL")
axes2.set_ylabel("Y2 LABEL")
axes2.set_title("title for graph2 between x and y")
```
